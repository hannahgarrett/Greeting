```python
import pygame
pygame.init()
from pygame import mixer

WHITE=(255,255,255)
BLACK=(0,0,0)
RED=(150,0,0)
GREEN=(0,150,0)
BLUE=(0,0,255)
BROWN = ( 160, 82, 45)
surface=pygame.display.set_mode((500,400))

surface.fill(RED)

pygame.display.set_caption("Greeting Card")

font=pygame.font.Font('freesansbold.ttf', 48)

#text surface
text=font.render("Season's Greetings!", True, RED, WHITE)

#create a rectangle object
textRect=text.get_rect()
X=500
y=100
textRect.center=(X//2,y//2)

points=[(225,150),(150,350), (300,350)]

pygame.draw.polygon(surface, GREEN, points)

pygame.draw.rect(surface, BROWN, pygame.Rect(200,350,50,50))

img=pygame.image.load("happyhol.png").convert()

resized_image=pygame.transform.scale(img, (200,150))

x=275 # X coordinate of the image
y=100 # Y coordinate of the image

surface.blit(resized_image, (x,y))

mixer.init()

mixer.music.load("jingle_bells.mp3")

mixer.music.set_volume(0.7)

mixer.music.play()

pygame.display.flip()

active=True
while active:
    
    surface.blit(text, textRect)
    
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            active=False
    pygame.display.update()
        
pygame.quit()
```
